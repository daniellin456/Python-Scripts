```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dens_Data = np.loadtxt("DensEoS.csv")
Temp_Data = np.loadtxt("TempEoS.csv")
```

```python
print(Dens_Data)
print( min(Dens_Data) )
print( max(Dens_Data) )
```

```python
print(Temp_Data)
print( min(Temp_Data) )
print( max(Temp_Data) )
```

```python
plt.figure( figsize=(12,10) )
plt.title(r"$\rho \; v.s \; T$", fontsize=18)
plt.xlabel(r"$log_{10}\rho$", fontsize=14)
plt.ylabel(r"$log_{10} T$", fontsize=14)
plt.hist2d(Dens_Data, Temp_Data, bins=(200,200) )
plt.colorbar()
plt.savefig("Barotropic_Table.png", dpi=500)
```
