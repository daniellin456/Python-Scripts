```python
import numpy as np
import matplotlib.pyplot as plt
from plot_style import *
plot_style()
```

```python
Dens_Data_Fill = np.fromfile("../resources/Dens_Table_Fill.bin").reshape( 1000, 625 )
Ener_Data_Fill = np.fromfile("../resources/Ener_Table_Fill.bin").reshape( 1000, 625 )
Temp_Data_Fill = np.fromfile("../resources/Temp_Table_Fill.bin").reshape( 1000, 625 )
Pres_Data_Fill = np.fromfile("../resources/Pres_Table_Fill.bin").reshape( 1000, 625 )
Cs_Data_Fill = np.fromfile("../resources/Cs_Table_Fill.bin").reshape( 1000, 625 )
```

```python
plt.figure( figsize=(20, 20))

plt.subplot(2, 2, 1 )
plt.title("Dens_Data_Fill")
plt.imshow( np.log10( Dens_Data_Fill ), origin = 'lower', cmap = plt.cm.jet )
plt.colorbar()

plt.subplot(2, 2, 2 )
plt.title("Ener_Data_Fill / Dens_Data_Fill")
plt.imshow( np.log10( Ener_Data_Fill / Dens_Data_Fill ), origin = 'lower', cmap = plt.cm.jet )
plt.colorbar()

plt.subplot(2, 2, 3 )
plt.title("Pres_Data_Fill")
plt.imshow( np.log10( Pres_Data_Fill ), origin = 'lower', cmap = plt.cm.Blues )
plt.colorbar()

plt.subplot(2, 2, 4 )
plt.title("Temp_Data_Fill")
plt.imshow( np.log10( Temp_Data_Fill ), origin = 'lower', cmap = plt.cm.Blues )
plt.colorbar()
```

```python
Dens_Data_Fill = Dens_Data_Fill.reshape( 625000 )
Ener_Data_Fill = Ener_Data_Fill.reshape( 625000 )
Pres_Data_Fill = Pres_Data_Fill.reshape( 625000 )
Temp_Data_Fill = Temp_Data_Fill.reshape( 625000 )

Zero_Index = np.where( Temp_Data_Fill == 0.0 )[0]
print("Length of Zero_Index: ", len( Zero_Index ) )
Dens_Data_Fill = np.delete( Dens_Data_Fill, Zero_Index )
Temp_Data_Fill = np.delete( Temp_Data_Fill, Zero_Index )
Ener_Data_Fill = np.delete( Ener_Data_Fill, Zero_Index )

Dens_Data_Fill = np.log10( Dens_Data_Fill )
Temp_Data_Fill = np.log10( Temp_Data_Fill )
Ener_Data_Fill = np.log10( Ener_Data_Fill )

plt.figure(figsize=(12, 6))
plt.subplot( 1, 2, 1 )
H, xedges, yedges = np.histogram2d(Dens_Data_Fill, Temp_Data_Fill, bins=30)
H = H.T
plt.imshow( H, origin = 'lower', cmap = plt.cm.jet, extent= [xedges[0], xedges[-1], yedges[0], yedges[-1]], aspect=3 )
plt.colorbar()

plt.subplot( 1, 2, 2 )
H, xedges, yedges = np.histogram2d(Ener_Data_Fill, Temp_Data_Fill, bins=30)
H = H.T
plt.imshow( H, origin = 'lower', cmap = plt.cm.jet, extent= [xedges[0], xedges[-1], yedges[0], yedges[-1]] )
plt.colorbar()
```
